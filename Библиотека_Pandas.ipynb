## Операция GroupBy

*  Шаг разбиения включает разделение на части и группировку объекта DataFrame на основе значений заданного ключа.
* Шаг применения включает вычисление какой-либо функции, обычно агрегирующей, преобразование или фильтрацию в пределах отдельных групп.
* На шаге объединения выполняется слияние результатов этих операций в выходной массив.

Добавим столбец с ключами

```python
dataframe['label'] = np.array(['A','A','B','C','A','B','A','C','B','C'])
dataframe
```

```python
dataframe.groupby('label')
```

```python
dataframe.groupby('label').sum()
```

### Индексация groupby

```python
dataframe.groupby('label')[['b','c']].sum()
```

## Агрегирование

Выполнение нескольких операций в рамках одной команды

```python
dataframe.groupby('label').aggregate(['min', max])
```

Для разных столбцов - своя агрегация

```python
group_result = dataframe.groupby('label').agg({'a': 'min',
                                      'b': 'max',
                                      'c': 'sum',
                                      'd': ['mean', 'std']
                                      })

group_result
```

Небольшая манипуляция с названиями столбцов

Чтобы они встали в одну строчку (без мульти-индекса)

```python
group_result.columns = [f"{i[0]}_{i[1]}" for i in group_result.columns]
group_result.reset_index(inplace=True)
group_result
```

# Объединение двух столбцов по ключу

`pd.merge(left, right, on, how)`

* `left` - левая таблица

* `on` - по каким стобцам делаете объединение

# Метод apply

Метод `.apply()` позволяет реализовывать более сложные преобразования над данными чем просто нахождение минимума или максимума

В перспективе мы ограничены только фантазией и знанием кода

В базовом варианте можно применять любую готовую функцию (например из библиотеки `numpy`)

```python
merge_results['a_log'] = merge_results['a'].apply(np.log)
merge_results
```

один из вариантов - использовать лямбда-функции языка python, для создания собственных преобразований

Суть преобразования - добавим к каждому элементу столбца 'a' единицу

```python
merge_results['a_new'] = merge_results['a'].apply(lambda x: x+1)
merge_results
```

можно вставлять условия в лямбда-функцию

Суть преобразования - вернём 0, если соответвующая ячейка в столбце 'b' меньше 0.5 или 1, если больше либо равна 0.5

```python
merge_results['b_new'] = merge_results['b'].apply(lambda x: 0 if x<0.5 else 1)
merge_results
```

можно работать сразу с несколькими столбцами

Суть преобразования - для каждой строки найдём максимум между значениями столбцов 'c' и 'd'

```python
merge_results['c_d_max'] = merge_results.apply(lambda x: max(x['c'],x['d']), axis = 1)
merge_results
```

можно подготовить отдельную функцию с более сложной логикой внутри, и даже использовать дополнительные аргументы

```python
def my_function(x, theshold, value):
    '''
    функция возвращает о, если значение x меньше порога theshold
    и возвращает  value, иначе (если значение x больше или равно порога theshold)
    '''
    if x< theshold:
        return 0
    else:
        return value
```

Суть преобразования - вернём 0, если соответвующая ячейка в столбце 'd' меньше среднего значения по группе или вернём значение стандартного оклонения , если больше либо равна 0.5 если соответвующая ячейка в столбце 'd' больше или равна среднему значению по группе

```python
merge_results['d_new'] = merge_results.apply(lambda x: my_function(x['d'], x['d_mean'], x['d_std'],), axis = 1)
merge_results
```

# Метод map

Для более простых преобразований (например замена значений по какому-то критерию) можно пользоваться методом `.map()`

В отличие от метода `.apply()`, который оперирует функциями,метод `.map()` может также работать со словарями

```python
dictionary = {0:'Zero', 1:'Один'}
```

Суть преобразования - если в столбце 'b_new' ячейка равна нулю, то мы пишем слово 'Zero', а если равна 1, то пишем слово 'Один'

```python
merge_results['b_cat'] = merge_results['b_new'].map(dictionary)
merge_results
```

Можно, как и в случае метода `.apply()` написать функцию с более сложными условиями для замены

```python
def replacer(x):
    if x<0.40:
        return 'Плохо'
    elif x<0.60:
        return 'Удовлетворительно'
    elif x<0.80:
        return 'Хорошо'
    else :
        return 'Отлично'
```

Суть преобразования - смотрим на столбец 'a' и выдаем строковую переменную в зависимости от значения

Замена имеет логику близкую к БРС, при переводе из 100-бальной системе в пяти-уровневую

```python
merge_results['a_cat'] = merge_results['a'].map(replacer)

merge_results
```
